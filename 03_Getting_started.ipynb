# Plotting a `sin` Wave with numpy
Let's start by importing the required modules.

```python
from math import sin,pi,cos
import numpy as np
import matplotlib.pyplot as plt
```

Let's see the `numpy` equivalent of a list array.

```python
list_array = [i for i in range(0,500,4)]
numpy_array1 = np.arange(0,500,4,dtype=int)
numpy_array2 = np.linspace(0,500,125,dtype=int,endpoint=False)
print("List Array : \n\n", list_array,"\n\n")
print("numpy Array using arange : \n\n",numpy_array1,"\n\n")
print("numpy Array using linspace : \n\n",numpy_array2)
```

let's make a sine wave with the numpy array and plot it.

```python
#amplitude of our wave 
a = 2

#frequency of out sine wave
f = 1                      

#angular frequency(omega) of our sine wave
w = 2*pi*f

#length of our sine wave in seconds
time = 2                                         

#we will take 1000 values of sine in one second,so our sampling freq is 1000 or sampling interval is 1ms
sampling_freq = 1000   

#T will have time*sampling_freq values in interval 0-2
T = np.linspace(0, time, time*sampling_freq)

#using sin function built in numpy to calculate the values
wave = a*np.sin(w*T)
plt.plot(T,wave)
plt.xlabel("t(ms)")
plt.ylabel("asin(wt)")
plt.title("Plot of asin(wt) ")
plt.show()
```

Let's create a more complex wave with both sin and cos components like `asin(wt)+acos(wt)`

```python
a = 2
f = 1                      
w = 2*pi*f
time = 2                                         
sampling_freq = 1000   
T = np.linspace(0, time, time*sampling_freq)
wave = a*np.sin(w*T)+a*np.cos(w*T)
plt.plot(T,wave)
plt.xlabel("t(ms)")
plt.ylabel("asin(wt)+acos(wt)")
plt.title("Plot of asin(wt)+acos(wt) ")
plt.show()
```

Let's recreate some examples from previous notebook.

```python
a = 2
f = 75                      
w = 2*pi*f
time = 1                                         
sampling_freq = 1000   
T1 = np.linspace(0, time, time*sampling_freq)
wave = a*np.sin(w*T1)+a*np.cos(w*T1)
plt.plot(T1,wave)
plt.xlabel("t(ms)")
plt.ylabel("asin(wt)+acos(wt)")
plt.title("Plot of asin(wt)+acos(wt) at 75Hz Frequency and 1000Hz Sampling Frequency")
plt.show()

                                       
sampling_freq = 125   
T2 = np.linspace(0, time,time*sampling_freq)
wave = a*np.sin(w*T2)+a*np.cos(w*T2)
plt.plot(T2,wave)
plt.xlabel("t(ms)")
plt.ylabel("asin(wt)+acos(wt)")
plt.title("Plot of asin(wt)+acos(wt) at 75Hz Frequency and 125Hz Sampling Frequency")
plt.show()
```

The plot of 125Hz doesn't look like the one in previous notebook, that is because np.linspace and lists behave differently. Let's see how.

```python
T1 = np.linspace(0, time,time*sampling_freq)
T2 = np.linspace(0,time*sampling_freq,time*sampling_freq,dtype=int)/sampling_freq

print("sin values for T1 \n")
for i in range(10):
    print(sin(w*T1[i]),"       ",T1[i])

print("\n\nsin values for T2 \n")
for i in range(10):
    print(sin(w*T2[i]),"       ",T2[i]/sampling_freq)
```

See the difference between `T1` and `T2` is of precision and that makes all the difference in the appearance of the sin wave. The position of sample and value is different than it would be in the case of a list. `T2` behaves the way our `lists` behaved and `T1` behaves as normal `np.linspace` should.
Here is a look at a `list`(ls) v/s normal `np.linspace`(npl) v/s `modified linspace`(mnpl).
for every sample ls and mnpl are same but npl is different hence the difference in the plot at 125Hz with `np.linspace`.


```python
ls = [i/sampling_freq for i in range(0,time*sampling_freq)] 
npl = np.linspace(0, time,time*sampling_freq)
mnpl = np.linspace(0,time*sampling_freq,time*sampling_freq,dtype=int)/sampling_freq
for i in range(10):
    print(ls[i],"::",npl[i],"::",mnpl[i])
```

`mnpl` looks very messy, let's find a replacement.

```python
l = np.arange(0,time,1/sampling_freq)
T = np.linspace(0,time*sampling_freq,time*sampling_freq,dtype=int)/sampling_freq
for i in range(10):
    print(l[i],T[i])
```

`np.arange` looks same as the `list` approach and modified `np.linspace` approach. let's see how it fares.
Let's plot the `asin(wt)+acos(wt)` again at 125Hz sampling frequency using modified linspace and arange that behaves like the list.

```python
print("Using modified np.linspace")
T = np.linspace(0,time*sampling_freq,time*sampling_freq,dtype=int)/sampling_freq
wave = a*np.sin(w*T)+a*np.cos(w*T)
plt.plot(T,wave)
plt.xlabel("t(ms)")
plt.ylabel("asin(wt)+acos(wt)")
plt.title("Plot of asin(wt)+acos(wt) at 75Hz Frequency and 125Hz Sampling Frequency")
plt.show()

with open("sin.txt",'w') as f:
    for i in wave:
        f.write(str(i)+"\n")
    f.close()

print("Using np.arange")
T = np.arange(0,time,1/sampling_freq)
wave = a*np.sin(w*T)+a*np.cos(w*T)
plt.plot(T,wave)
plt.xlabel("t(ms)")
plt.ylabel("asin(wt)+acos(wt)")
plt.title("Plot of asin(wt)+acos(wt) at 75Hz Frequency and 125Hz Sampling Frequency")
plt.show()
```

Now we can replicate the sin wave using numpy with a little bit cleaner code and more optimized professional/scientific tools. But the question "how does the same signal look different at different `sampling_freq`?" still stands. we will explore that in the next section called **Nyquist Theorem**.
